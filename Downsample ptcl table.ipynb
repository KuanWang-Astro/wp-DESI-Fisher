```python
%pylab inline
from halotools.sim_manager import CachedHaloCatalog
```

```python
halocat = CachedHaloCatalog(simname='bolplanck',redshift=0,halo_finder='rockstar',version_name='halotools_v0p4')
```

```python
rate = 0.1
```

```python
mask = np.random.rand(len(halocat.ptcl_table))<rate
```

```python
halocat.ptcl_table[mask]
```

```python
from halotools.mock_observables import return_xyz_formatted_array
```

```python
table = halocat.ptcl_table[mask]
ptclpos = return_xyz_formatted_array(*(table[ax] for ax in 'xyz'), period=250)
```

```python
ptclpos
```

```python
np.savetxt('bolplanck_1of10_ptcl.txt',ptclpos)
```

```python
plt.plot(ptclpos[:,1],ptclpos[:,2],'.',markersize=0.1)
```
