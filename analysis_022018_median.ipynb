```python
c = np.zeros((8,7))
for i in range(7):
    c[:,i] = np.argsort(bolp_19p0.constraint[:,i])
```

```python
def rgb(c):
    if c>2:
        return np.array((1,c*0.25-0.75,c*0.25-0.75))
    else:
        return np.array((0.4+0.2*c,0,0))
```

```python
crgb = np.ones((8,7,3))
```

```python
for i in range(8):
    for j in range(7):
        crgb[i,j] = rgb(c[i,j])
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111, frameon=False, xticks=[], yticks=[])
plt.table(cellText=bolp_19p0.constraint,cellColours=crgb,rowLabels=comb_list,colLabels=param_list,loc='center')
```
