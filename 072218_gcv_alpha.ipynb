```python
npylist = ['gcv_alpha','gcv_logM1','gcv_sigmalogM','gcv_logM0','gcv_logMmin','gcv_Acen','gcv_Asat']
```

```python
alphas = np.linspace(0.05,0.9,18)
```

```python
test = gcv_236[npylist[0]]
for i in range(100):
    plt.figure()
    #plt.semilogy()
    plt.plot(1/alphas,test[i],'.-')
    plt.title(i)
    plt.show()
```
