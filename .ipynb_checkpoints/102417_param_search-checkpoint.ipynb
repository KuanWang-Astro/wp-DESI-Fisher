```python
param_median = param[0]
param_median
```

```python
mask_median = np.logical_and(param.T[0]==param_median[0],param.T[1]==param_median[1])
for i in range(5):
    mask_median = np.logical_and(mask_median,param.T[i+2]==param_median[i+2])
```

```python
covar_real = np.cov(wp[mask_median].T)

covar_cosmic = np.mean(wpcov[mask_median],axis=0)

covar_tot = covar_cosmic+covar_real
```

```python
n_med = np.sum(mask_median)
```

```python
def dp(i):
    dp = param[i]-param_median
    return np.where(dp!=0)[0][0],dp[np.where(dp!=0)]   ##i,dpi
```

```python
didpi = np.zeros((len(ngal)-n_med,2))
```

```python
for i in range(n_med,len(ngal)):
    didpi[i-n_med,0] = dp(i)[0]
    didpi[i-n_med,1] = dp(i)[1]
```

```python
dtype = [('original_idx', int), ('i', int), ('dpi', float)]
values = zip(range(0,len(ngal)-n_med),didpi[:,0],didpi[:,1])
d_param = np.array(values, dtype=dtype)
```

```python
mask_alpha = np.logical_and(d_param['i']==0,d_param['dpi']>-0.2)
mask_logM1 = d_param['i']==1
mask_sigmalogM = d_param['i']==2
mask_logM0 = d_param['i']==3
mask_logMmin = d_param['i']==4
mask_Acen = np.logical_and(d_param['i']==5,d_param['dpi']<1-param_median[5])
mask_Asat = np.logical_and(d_param['i']==6,d_param['dpi']<1-param_median[6])
```

```python
def dfdp(i,func=wp):
    d_func = np.mean(func[mask_median],axis=0)-func[i]
    dp = param[i]-param_median
    return d_func/np.sum(dp),np.where(dp!=0)[0][0],np.sum(dp)   ##df/dpi,i,dpi
```

```python
def Fisher(i,j,m,n,func=wp,covar=covar_tot):
    Iij = np.dot(dfdpi[np.where(dfdpi[:,0]==i)][m,2:],np.dot(np.linalg.inv(covar),dfdpi[np.where(dfdpi[:,0]==j)][n,2:]))
    return Iij
```

```python
def Fmatrix(d_vector):
    fmatrix = np.zeros((7,7))
    dp = np.zeros(7)
    for i in range(7):
        dp[i] = dfdpi[np.where(dfdpi[:,0]==i)][d_vector[i],1]
        for j in range(7):
            fmatrix[i,j] = Fisher(i,j,d_vector[i],d_vector[j])
    return fmatrix,dp
```
