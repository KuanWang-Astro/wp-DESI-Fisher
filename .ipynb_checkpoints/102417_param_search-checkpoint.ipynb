```python
mask_median = np.logical_and(param.T[0]==param_median[0],param.T[1]==param_median[1])
for i in range(5):
    mask_median = np.logical_and(mask_median,param.T[i+2]==param_median[i+2])
```

```python
covar_real = np.cov(wp[mask_median].T)

covar_cosmic = np.mean(wpcov[mask_median],axis=0)

covar_tot = covar_cosmic+covar_real
```

```python
n_med = np.sum(mask_median)
```

```python
def dp(i):
    dp = param[i]-param_median
    return np.where(dp!=0)[0][0],dp[np.where(dp!=0)]   ##i,dpi
```

```python
didpi = np.zeros((len(ngal)-n_med,2))
```

```python
for i in range(n_med,len(ngal)):
    didpi[i-n_med,0] = dp(i)[0]
    didpi[i-n_med,1] = dp(i)[1]
```

```python
dfdp = np.zeros((7,19))
```

```python
def dfdpi(ip,mask,ir):
    return np.linalg.lstsq((param[n_med+d_param['original_idx'][mask],ip]-param_median[ip])[:,np.newaxis],\
                           wp[n_med+d_param['original_idx'][mask],ir]-np.mean(wp[mask_median][:,ir],axis=0))[0]
```

```python
mask_list = [mask_alpha,mask_logM1,mask_sigmalogM,mask_logM0,mask_logMmin,mask_Acen,mask_Asat]
```

```python
for i,mask in enumerate(mask_list):
    for j in range(19):
        dfdp[i,j] = dfdpi(i,mask,j)
```

```python
fmatrix = np.zeros((7,7))
for i in range(7):
    for j in range(7):
        fmatrix[i,j] = np.dot(dfdp[i],np.dot(np.linalg.inv(covar_tot),dfdp[j]))
```
